```python
import pandas as pd
import datetime as dt

data_dict = {
    "Datetime" : [dt.datetime(2022,2,15, 9, 35+i, 0) for i in range(10)],
    "Close": [100 + 0.01 * i for i in range(10)],
}
df = pd.DataFrame(data_dict)

print(df)

other_data_dict = {
    "Datetime" : [dt.datetime(2022,2,15, 9, 40, 0) + dt.timedelta(minutes=i) for i in range(10)],
    "Close": [100.05 + 0.01 * i for i in range(10)],
    "Indicator": [100 + 0.01 * i for i in range(10)]
}
other_df = pd.DataFrame(other_data_dict)
print(other_df)
```

```python
df.merge(other_df, how="left")
```

```python
df.merge(other_df, how="right")
```

```python
df.merge(other_df, how="inner")
```

```python
df.merge(other_df, how="outer")
```
